## A- Contour Features - Star image:

```python
import cv2 as cv
import matplotlib.pyplot as plt 

star = cv.imread('star.jpg')
star2 = cv.imread('star2.png')
star3 = cv.imread('star3.png')

assert star is not None, "star.jpg could not be read, check with os.path.exists()"
assert star2 is not None, "star2.png could not be read, check with os.path.exists()"
assert star3 is not None, "star3.png could not be read, check with os.path.exists()"

#for star
imgray = cv.imread('star.jpg', cv.IMREAD_GRAYSCALE)

ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

largest_contour = max(contours, key=cv.contourArea)

epsilon_1percent = 0.01 * cv.arcLength(largest_contour, True)
approx_1percent = cv.approxPolyDP(largest_contour, epsilon_1percent, True)

star_with_contour = star.copy()
cv.drawContours(star_with_contour, [approx_1percent], 0, (0, 0, 255), 10) 


epsilon_10percent = 0.1 * cv.arcLength(largest_contour, True)
approx_10percent = cv.approxPolyDP(largest_contour, epsilon_10percent, True)
star_with_contour10 = star.copy()
cv.drawContours(star_with_contour10, [approx_10percent], 0, (0, 0, 255), 10) 

#for star2
imgray_star2 = cv.cvtColor(star2, cv.COLOR_BGR2GRAY)
ret_star2, thresh_star2 = cv.threshold(imgray_star2, 127, 255, 0)
contours_star2, hierarchy_star2 = cv.findContours(thresh_star2, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
largest_contour_star2 = max(contours_star2, key=cv.contourArea)

epsilon_1percent_star2 = 0.01 * cv.arcLength(largest_contour_star2, True)
approx_1percent_star2 = cv.approxPolyDP(largest_contour_star2, epsilon_1percent_star2, True)

star2_with_contour = star2.copy()
cv.drawContours(star2_with_contour, [approx_1percent_star2], 0, (0, 0, 255), 10)

epsilon_10percent_star2 = 0.1 * cv.arcLength(largest_contour_star2, True)
approx_10percent_star2 = cv.approxPolyDP(largest_contour_star2, epsilon_10percent_star2, True)

star2_with_contour10 = star2.copy()
cv.drawContours(star2_with_contour10, [approx_10percent_star2], 0, (0, 0, 255), 10)

#for star3
imgray_star3 = cv.cvtColor(star3, cv.COLOR_BGR2GRAY)
ret_star3, thresh_star3 = cv.threshold(imgray_star3, 127, 255, 0)
contours_star3, hierarchy_star3 = cv.findContours(thresh_star3, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
largest_contour_star3 = max(contours_star3, key=cv.contourArea)

epsilon_1percent_star3 = 0.01 * cv.arcLength(largest_contour_star3, True)
approx_1percent_star3 = cv.approxPolyDP(largest_contour_star3, epsilon_1percent_star3, True)

star3_with_contour = star3.copy()
cv.drawContours(star3_with_contour, [approx_1percent_star3], 0, (0, 0, 255), 10)

epsilon_10percent_star3 = 0.1 * cv.arcLength(largest_contour_star3, True)
approx_10percent_star3 = cv.approxPolyDP(largest_contour_star3, epsilon_10percent_star3, True)

star3_with_contour10 = star3.copy()
cv.drawContours(star3_with_contour10, [approx_10percent_star3], 0, (0, 0, 255), 10)

plt.subplot(331), plt.imshow(cv.cvtColor(star, cv.COLOR_BGR2RGB)), plt.title('Star'), plt.axis('off')
plt.subplot(332), plt.imshow(cv.cvtColor(star_with_contour, cv.COLOR_BGR2RGB)), plt.title('Star, ε = 1%'), plt.axis('off')
plt.subplot(333), plt.imshow(cv.cvtColor(star_with_contour10, cv.COLOR_BGR2RGB)), plt.title('Star, ε = 10%'), plt.axis('off')

plt.subplot(334), plt.imshow(cv.cvtColor(star2, cv.COLOR_BGR2RGB)), plt.title('Star 2'), plt.axis('off')
plt.subplot(335), plt.imshow(cv.cvtColor(star2_with_contour, cv.COLOR_BGR2RGB)), plt.title('Star 2, ε = 1%'), plt.axis('off')
plt.subplot(336), plt.imshow(cv.cvtColor(star2_with_contour10, cv.COLOR_BGR2RGB)), plt.title('Star 2, ε = 10%'), plt.axis('off')

plt.subplot(337), plt.imshow(cv.cvtColor(star3, cv.COLOR_BGR2RGB)), plt.title('Star 3'), plt.axis('off')
plt.subplot(338), plt.imshow(cv.cvtColor(star3_with_contour, cv.COLOR_BGR2RGB)), plt.title('Star 3, ε = 1%'), plt.axis('off')
plt.subplot(339), plt.imshow(cv.cvtColor(star3_with_contour10, cv.COLOR_BGR2RGB)), plt.title('Star 3, ε = 10%'), plt.axis('off')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

## B- Contour Properties - Hindian_map image:

```python
import numpy as np

img = cv.imread('hindian_map.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"

ret,thresh = cv.threshold(img,127,255,0)
contours,hierarchy = cv.findContours(thresh, 1, 2)

largest_contour = max(contours, key=cv.contourArea)

contour_img = np.zeros_like(img)
cv.drawContours(contour_img, [largest_contour], 0, 255, 2)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Image Originale'), plt.axis('off')
plt.subplot(122), plt.imshow(contour_img, cmap='gray'), plt.title('Contour'), plt.axis('off')

plt.show()

cnt = largest_contour
M = cv.moments(cnt)
print( M )

#centroids
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])

print("\ncx :", cx, " cy : ", cy)

#area
print("\nArea ",  M['m00'])

#perimeter
perimeter = cv.arcLength(cnt,True)
print("\nPerimeter :",  perimeter)

#aspect ratio
x,y,w,h = cv.boundingRect(cnt)
aspect_ratio = float(w)/h
print("\nAspect ratio :", aspect_ratio)

#extent
x,y,w,h = cv.boundingRect(cnt)
rect_area = w*h
extent = float(M['m00'])/rect_area
print("\nExtent :", extent)

#Solidity
area = cv.contourArea(cnt)
hull = cv.convexHull(cnt)
hull_area = cv.contourArea(hull)
solidity = float(area)/hull_area
print("\nSolidity :", solidity)

#Equivalent Diameter 
area = cv.contourArea(cnt)
equi_diameter = np.sqrt(4*area/np.pi)
print("\nEquivalent Diameter : ", equi_diameter)

#Orientation
#(x,y),(MA,ma),angle = cv.fitEllipse(cnt)
#print("\nCoordinates of the  Ellipse center (x,y) : ", x," , ",y)

#masks & pixels points
mask = np.zeros(img.shape,np.uint8)
cv.drawContours(mask,[largest_contour],0,255,-1)
pixelpoints = np.transpose(np.nonzero(mask))
#pixelpoints = cv.findNonZero(mask)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('hindian_map'), plt.axis('off')
plt.subplot(122), plt.imshow(mask, cmap='gray'), plt.title('Mask'), plt.axis('off')

plt.show()

image_copy = img.copy()

for point in pixelpoints:
    cv.circle(image_copy, tuple(point[::-1]), 1, 255, -1)  

plt.figure()
plt.imshow(image_copy, cmap='gray')
plt.title('Image with Mask Points')
plt.axis('off')

plt.show()

min_val, max_val, min_loc, max_loc = cv.minMaxLoc(img,mask = mask)
print("Minimum Value:", min_val)
print("Maximum Value:", max_val)
print("Location of Minimum Value:", min_loc)
print("Location of Maximum Value:", max_loc)

mean_val = cv.mean(img,mask = mask)
print("\nMean Value:", mean_val)
```

## C- Contours: More Functions

```python
img = cv.imread('star2.png')

assert img is not None, "file could not be read, check with os.path.exists()"

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(img_gray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, 2, 1)

cnt = max(contours, key=cv.contourArea)

hull = cv.convexHull(cnt, returnPoints=False)
# Find convexity defects in the contour and convex hull
defects = cv.convexityDefects(cnt, hull)

if defects is not None:
    for i in range(defects.shape[0]):
        s, e, f, d = defects[i, 0]
        start = tuple(cnt[s][0])
        end = tuple(cnt[e][0])
        far = tuple(cnt[f][0])
        
        # Draw a line connecting the start and end points of the defect
        cv.line(img, start, end, [0, 255, 0], 2)
        # Draw a circle at the far point of the defect
        cv.circle(img, far, 5, [0, 0, 255], -1)
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title('Image with Convexity Defects')
    plt.show()
else:
    print('No convexity defects found.')
```
